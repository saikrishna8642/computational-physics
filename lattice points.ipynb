```python
import numpy as np
import math as m
from matplotlib import pyplot as plt
class Lattice:
       
    def square():
        a = int(input("\nEnter the side of the lattice:\t"))
        x = int(input("\nEnter the point of study:\t"))
        i = 0
        j = 0
        n = 0
        for o in range(1,a+1):
            n += a
        q = np.array([])
        p = [0,1]
        c = [1,0]
        for i in range(0,a):
            y = np.zeros((a,2))
            b = np.multiply(p,i)
            y[0][0] = np.add(y[0][0],b[0])
            y[0][1] = np.add(y[0][1],b[1])
            q = np.append(q,y[0])
            for j in range(0,a-1):
                y[j+1][0] = np.add(y[j][0],c[0])
                y[j+1][1] = np.add(y[j][1],c[1])
                q = np.append(q,y[j+1])
        q = np.reshape(q,(-1,2))
        d = np.array([])
        f = np.array([])
        g = x-1
        for i in range(0,n):
            e = abs(np.subtract(q[i][0],q[g][0]))
            h = abs(np.subtract(q[i][1],q[g][1]))
            if e == 1 and h == 0:
                d = np.append(d,q[i])
                f = np.append(f,i+1)
            if e == 0 and h == 1:
                d = np.append(d,q[i])
                f = np.append(f,i+1)
            if e == (a-1) and h == 0:
                d = np.append(d,q[i])
                f = np.append(f,i+1)
            if e == 0 and h == (a-1):
                d = np.append(d,q[i])
                f = np.append(f,i+1)
        d = np.reshape(d,(-1,2))
        print("\nThe nearest neighbours of\t",x,"\twith coordinates\t",q[g],"\tare:")
        for i in range(0,len(f)):
            print("\n",f[i],"\n",d[i])
        h = np.zeros(n)
        k = np.zeros(n)
        for i in range(0,n):
            h[i] = q[i][0]
            k[i] = q[i][1]
        plt.plot(h,k,'.')
        plt.show()
        return 
    def triangle():
        a = int(input("\nEnter the number of points on each side:\t"))
        x = int(input("\nEnter the point of study:\t"))
        i = 0
        j = 0
        n = 0
        for o in range(0,a):
            n += a
    
        p = [1/2,m.sqrt(3)/2]
        c = [1,0]
        q = np.array([])
        for i in range(0,a):
            y = np.zeros((a,2))
            b = np.multiply(p,i)
            y[0][0] = np.add(y[0][0],b[0])
            y[0][1] = np.add(y[0][1],b[1])
            q = np.append(q,y[0])
            for j in range(0,a-1):
                y[j+1][0] = np.add(y[j][0],c[0])
                y[j+1][1] = np.add(y[j][1],c[1])
                q = np.append(q,y[j+1])
        q = np.reshape(q,(-1,2))
        i = 0
        d = np.array([])
        f = np.array([])
        g = x-1
        i = 0
        l = 0
        h = 0
        for i in range(0,n):
            z = np.zeros(2)
            l = abs(np.subtract(q[g][0],q[i][0]))
            h = abs(np.subtract(q[g][1],q[i][1]))
            z[0] = l
            z[1] = h
            e = round(pow(z[0],2) + pow(z[1],2),3)
            if e == 1 or e == pow(a-1,2) or (l == 3 and h == 4*p[1]):
                f = np.append(f,i+1)
                d = np.append(d,q[i])
        d = np.reshape(d,(-1,2))
        print("\nThe nearest neighbours of\t",x,"\twith coordinates\t",q[g],"are:")
        for i in range(0,len(f)):
            print("\n",f[i],"\n",d[i])
        h = np.zeros(n)
        k = np.zeros(n)
        for i in range(0,n):
            h[i] = q[i][0]
            k[i] = q[i][1]
        plt.plot(h,k,'.')
        plt.show()
        return
    def honeycomb():
        a = int(input("\nEnter the number of points on each side:\n"))
        x = int(input("\nEnter the point of study:\t"))
        i = 0
        j = 0
        n = 0
        n = a*a
        p = [1/2,m.sqrt(3)/2]
        c = [1,0]
        b = [2,0]
        q = np.array([])
        y = np.zeros((a,2))
        for i in range(0,a):
            if i != 0:
                if i%2 == 1:
                    y[0][0] = np.add(y[0][0],p[0])
                    y[0][1] = np.add(y[0][1],p[1])
                else:
                    y[0][0] = abs(np.subtract(y[0][0],p[0]))
                    y[0][1] = np.add(y[0][1], p[1])
            j = 0
            while j < a:
                if j != a-1:
                    if (i%2 == 0 and j%2 == 0)or(i%2 == 1 and j%2 == 1) :
                        y[j+1][0] = np.add(y[j][0],b[0])
                        y[j+1][1] = np.add(y[j][1],b[1])
                    elif (i%2 == 0 and j%2 == 1)or(i%2 == 1 and j%2 == 0):
                        y[j+1][0] = np.add(y[j][0],c[0])
                        y[j+1][1] = np.add(y[j][1],c[1])                        
                else:
                    break;
                q = np.append(q,y[j])
                j+=1
            q = np.append(q,y[j])
        q = np.reshape(q,(-1,2))
        i = 0
        d = np.array([])
        z = np.zeros(2)
        f = np.array([])
        g = x-1
        b = round((m.sqrt(3)*float((a-1)/2)),4)
        l = int(a/2+a-1)
        for i in range(0,n):
            z = abs(np.subtract(q[g],q[i]))
            e = round(float(pow(z[0],2) + pow(z[1],2)),5)
            z[1] = round(z[1],4)
            if (e == 1) or (z[0] == 0.5 and z[1] == b) or (e == pow(l,2)):
                f = np.append(f,i+1)
                d = np.append(d,q[i])
        d = np.reshape(d,(-1,2))
        print("\nThe nearest neighbours of\t",x,"\twith coordinates\t",q[g],"are:")
        for i in range(0,len(f)):
            print("\n",f[i],"\n",d[i])
        h = np.zeros(n)
        k = np.zeros(n)
        for i in range(0,n):
            h[i] = q[i][0]
            k[i] = q[i][1]
        plt.plot(h,k,'.')
        plt.show()
        return
```
